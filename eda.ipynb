```python
df['spring_min_temp'] = df[['march_min_temp', 'april_min_temp', 'may_min_temp']].min(axis=1)
df['spring_avg_temp'] = df[['march_avg_temp', 'april_avg_temp', 'may_avg_temp']].mean(axis=1)
df['spring_max_temp'] = df[['march_max_temp', 'april_max_temp', 'may_max_temp']].max(axis=1)

df['summer_min_temp'] = df[['june_min_temp', 'july_min_temp', 'august_min_temp',]].min(axis=1)
df['summer_avg_temp'] = df[['june_avg_temp', 'july_avg_temp', 'august_avg_temp',]].mean(axis=1)
df['summer_max_temp'] = df[['june_max_temp', 'july_max_temp', 'august_max_temp',]].max(axis=1)

df['fall_min_temp'] = df[['september_min_temp', 'october_min_temp', 'november_min_temp']].min(axis=1)
df['fall_avg_temp'] = df[['september_avg_temp', 'october_avg_temp', 'november_avg_temp']].mean(axis=1)
df['fall_max_temp'] = df[['september_max_temp', 'october_max_temp', 'november_max_temp']].max(axis=1)

df['winter_min_temp'] = df[['december_min_temp', 'february_min_temp', 'march_min_temp']].min(axis=1)
df['winter_avg_temp'] = df[['december_avg_temp', 'february_avg_temp', 'march_avg_temp']].mean(axis=1)
df['winter_max_temp'] = df[['december_max_temp', 'february_max_temp', 'march_max_temp']].max(axis=1)
```

The average temperature of state 1 ranges from about `58 - 68`. Also, with increasing temperature, the wind speed decreases. Accordingly, we can form two clusters based off the average temperature and assign the mean max wind speed for temperatures falling in each bucket.

Clusters:
* `avg_temp` > 62 have `max_wind_speed` = 12.5
* `avg_temp` < 62 have `max_wind_speed` = 17

```python
df.loc[(df.max_wind_speed.isnull()) & (df.State_Factor == 'State_1') & (df.avg_temp > 62), 'max_wind_speed'] = 12.5
df.loc[(df.max_wind_speed.isnull()) & (df.State_Factor == 'State_1') & (df.avg_temp < 62), 'max_wind_speed'] = 17
```
